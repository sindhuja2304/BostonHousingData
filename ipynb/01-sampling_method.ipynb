```r
BostonURL <- "https://archive.ics.uci.edu/ml/machine-learning-databases/housing/housing.data"
Boston <- read.csv(url(BostonURL), sep = "", header = FALSE)
```

```r
colnames(Boston) <- c('Crime', 'LrgLots', 'IndAcr', 'River', 'NOX', 'Rms', 'OwnOcc', 'DistWork', 'HiwayAcc', 'PropTax', 'EdRat', 'Min', 'LowStatus', 'MedVal')
```

```r
BostonNum <- Boston
BostonNum <- subset(BostonNum, select = c('Crime', 'LrgLots', 'IndAcr', 'NOX', 'Rms', 'OwnOcc', 'DistWork', 'PropTax', 'EdRat', 'Min', 'LowStatus', 'MedVal'))
```

```r
summary(BostonNum)
str(BostonNum)
```

##### For the sampling we have taken the sample sizes as n = 5, n=30, n=60, n=80 and n=100.

```r
set.seed(101)
BostSamp05 <- BostonNum[sample(1:nrow(BostonNum), 05, replace = FALSE),]
BostSamp80 <- BostonNum[sample(1:nrow(BostonNum), 80, replace = FALSE),]
BostSamp100 <- BostonNum[sample(1:nrow(BostonNum), 100, replace = FALSE),]
```

```r
BMn <- map_dbl(BostonNum, mean)
BSd <- map_dbl(BostonNum, sd)
BMd <- map_dbl(BostonNum, median)
BSk <- map_dbl(BostonNum, skewness)
BKu <- map_dbl(BostonNum, kurtosis)
BStatsDF <- rbind(BMn, BSd, BMd, BSk, BKu)
BStatsDF
```

#### When you compare the below density charts n = 100 looks more like the original data set.

```r
plot_density(BostonNum, title = 'Original')
plot_density(BostSamp05, title = 'n = 05')
plot_density(BostSamp80, title = 'n = 80')
plot_density(BostSamp100, title = 'n = 100')
```

#### When comparing the data in tabular format you can see that n=100 has approximately the same values as the original data set.

```r
BSamp05Mn <- map_dbl(BostSamp05, mean)
BSamp80Mn <- map_dbl(BostSamp80, mean)
BSamp100Mn <- map_dbl(BostSamp100, mean)

BSamp05Sd <- map_dbl(BostSamp05, sd)
BSamp80Sd <- map_dbl(BostSamp80, sd)
BSamp100Sd <- map_dbl(BostSamp100, sd)

BSamp05Md <- map_dbl(BostSamp05, median)
BSamp80Md <- map_dbl(BostSamp80, median)
BSamp100Md <- map_dbl(BostSamp100, median)

BSamp05Sk <- map_dbl(BostSamp05, skewness)
BSamp80Sk <- map_dbl(BostSamp80, skewness)
BSamp100Sk <- map_dbl(BostSamp100, skewness)

BSamp05Ku <- map_dbl(BostSamp05, kurtosis)
BSamp80Ku <- map_dbl(BostSamp80, kurtosis)
BSamp100Ku <- map_dbl(BostSamp100, kurtosis)
```

```r
BStatsDF
BSamp05DF <- rbind(BSamp05Mn, BSamp05Sd, BSamp05Md, BSamp05Sk, BSamp05Ku)
BSamp80DF <- rbind(BSamp80Mn, BSamp80Sd, BSamp80Md, BSamp80Sk, BSamp80Ku)
BSamp100DF <- rbind(BSamp100Mn, BSamp100Sd, BSamp100Md, BSamp100Sk, BSamp100Ku)
```

```r
as_tibble(BStatsDF, rownames = 'Statistic')
as_tibble(BSamp05DF, rownames = 'Statistic')
as_tibble(BSamp80DF, rownames = 'Statistic')
as_tibble(BSamp100DF, rownames = 'Statistic')
```

#### Inititally we take different sample sizes based on the survey monkey sample calculator.We get the sample sizes n=5,n=30,n=80 and n=100.When we compare the density plots of all of the sample only n=80 and n=100 closely resemble the density plot of original data set.
